```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
data=pd.read_csv(r"D:\Credit_Card_Fraud_Detection\notebooks\data\creditcard.csv")
```

```python
legit = data[data["Class"]==0]
fraud = data[data["Class"]==1]
```

```python
sns.set_style('darkgrid')
sns.set_palette('mako')
sns.distplot(x = fraud['Time'],label = 'Time')
```

```python
legit_sample = legit.sample(n=492)
new_data = pd.concat((legit_sample,fraud),axis=0)
```

```python
sns.set_palette('PiYG')
sns.displot(data = new_data,x = 'Time')
```

```python
sns.set_palette('cool')
sns.displot(data = data,x = 'Time')
```

```python
sns.set_palette('brg_r')
sns.scatterplot(x = fraud['Time'],y = fraud['Amount'],data = new_data)
```

```python
sns.set_palette('bwr')
sns.scatterplot(x = 'Time',y = 'Amount',data = new_data)
```
